```python
corellation_matrix = df.drop(columns=['year']).corr(numeric_only= True)

# Create heatmap
sns.heatmap(corellation_matrix, annot=True)
plt.title('Correlation matrix for numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
corr_pairs = corellation_matrix.unstack()
corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```

```python
# To check if the company has any corellation with gross income, we will perform the anova test
# Hypothesis:
# Ho:  There is no significant difference in the mean gross revenue among all the companies.
# H1:At least one of the companies has a significantly different mean gross revenue compared to the others.
from scipy.stats import f_oneway
# Group the gross revenue data by company
grouped_data = [group['gross'] for name, group in df.groupby('company')]
# Perform one-way ANOVA
anova_result = f_oneway(*grouped_data)

# Print ANOVA result
print("ANOVA Result:")
print("F-statistic:", anova_result.statistic)
print("p-value:", anova_result.pvalue)
```

```python
alpha = 0.05  # Chosen significance level
if anova_result.pvalue <= alpha:
    print("Reject the null hypothesis. The result is statistically significant.")
else:
    print("Fail to reject the null hypothesis. The result is not statistically significant.")
```

```python
# We can say that the significant difference in gross revenue among companies suggests that movies produced 
# by certain companies tend to make more money compared to movies produced by other companies.
```
